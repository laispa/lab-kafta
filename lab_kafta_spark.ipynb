```python
subreddit = reddit.subreddit("all")

# Coletar até 100 postagens que mencionam a palavra "Trump"
posts = subreddit.search("Trump", limit=100, time_filter="week")  # "day", "hour", ou "week"
```

```python
from transformers import pipeline

# Carregar os pipelines de análise de sentimentos, classificação de texto e tradução
sentiment_analysis = pipeline("sentiment-analysis")
text_classification = pipeline("zero-shot-classification")
```

```python

# Função para fazer análise de sentimentos
def analisar_sentimento(texto):
    resultado = sentiment_analysis(texto)
    return resultado[0]  # Retorna o sentimento e a confiança (ex: POSITIVE ou NEGATIVE)

# Função para classificar o texto
def classificar_texto(texto):
    # Definindo possíveis categorias
    labels = ["politics", "sports", "technology", "entertainment", "health"]
    resultado = text_classification(texto, candidate_labels=labels)
    return resultado['labels'][0], resultado['scores'][0]  # Retorna a categoria com maior pontuação
```

```python

for post in posts:
    # Analisando o sentimento
    sentiment = analisar_sentimento(post.title)

    # Classificando o texto
    categoria, score = classificar_texto(post.title)

    message = {
        'title': post.title,
        'url': post.url,
        'author': post.author.name if post.author else 'Unknown',
        'sentiment': sentiment,
        'category': categoria,
        'category_score': score
    }

    # Imprimir o conteúdo do sentiment, classificação e tradução
    print(f"Sentimento do post '{post.title}': {sentiment}")
    print(f"Categoria do post '{post.title}': {categoria} (Score: {score})")

    producer.send('canalinput', value=message)

print("Dados com análise de sentimentos, classificação e tradução enviados para o Kafka com sucesso!")

```

```python
from pyspark.sql import SparkSession
```

```python
SPARK_MASTER="spark://localhost:7077"
APP_NAME="Lab Kafka Niajus"

KAFKA_HOST="localhost:9092"
KAFKA_TOPIC_IN="canalinput"
KAFKA_TOPIC_OUT="canaloutput"

INTERVAL = "5 seconds"
```

```python
spark = SparkSession.builder.master(SPARK_MASTER).appName(APP_NAME).getOrCreate()
```

```python
from pyspark.sql.functions import explode, split, col, upper, window, to_json, struct, lit
```

```python
words = messages \
            .select(
                explode(split(col("value"), "\s+")).alias("word"),
                messages.timestamp
        ).select (
            upper(col("word")).alias("word"),
            col("timestamp")
        )
```

```python
wordCounts = words.withWatermark("timestamp", INTERVAL) \
                .groupBy(
                    window(words.timestamp, INTERVAL, INTERVAL), "word"
                ).count()
```

```python
wordCountsJson = wordCounts.select(
    lit("1").alias("key"),
    to_json(struct("word", "count", 'window.start')).alias("value"))
```

```python
wck = wordCountsJson.writeStream.outputMode("complete").format("kafka").option("kafka.bootstrap.servers", KAFKA_HOST) \
        .option("topic", KAFKA_TOPIC_OUT).option("checkpointLocation", "/tmp/spark-kafka") \
        .trigger(processingTime=INTERVAL).start()
```
